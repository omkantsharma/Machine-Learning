```python
x = df.iloc[:,:-1] # x = dataframe
y = df.iloc[:,-1]  # y = series
print(x.shape)
print(y.shape)
```

```python
x.columns
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
model = DecisionTreeRegressor(max_depth=4)
# try building model with different values of max_depth and then visualize its corresponding tree
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
